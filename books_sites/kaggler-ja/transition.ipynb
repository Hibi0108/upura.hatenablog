```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
```

```python
datetime = [
    '2017-08-22',
    '2017-09-22',
    '2017-10-22',
    '2017-11-22',
    '2017-12-22',
    '2018-01-22',
    '2018-02-22',
    '2018-03-22',
    '2018-04-22',
    '2018-05-22',
    '2018-06-22',
    '2018-07-22',
    '2018-08-22',
    '2018-08-26',
    '2018-08-31',
    '2018-09-04',
    '2018-09-13',
    '2018-09-25',
    '2018-10-09',
    '2018-10-27',
    '2018-11-03'
]

num = [
    0,
    400,
    450,
    550,
    780,
    880,
    1000,
    1100,
    1200,
    1400,
    1850,
    2200,
    2400,
    2500,
    2600,
    2700,
    2800,
    2900,
    3000,
    3200,
    3300
]
```

```python
df = pd.DataFrame({'datetime': datetime, 'num': num})
```

```python
df['datetime'] = pd.to_datetime(df['datetime'])
```

```python
import datetime
df['diff_datetime'] = [(i - datetime.datetime(2017, 8, 22)).days for i in df['datetime']]
```

```python
df
```

```python
plt.rcParams["font.size"] = 25
fig = plt.figure(figsize=(20,10))
plt.xlabel("year-month")
plt.ylabel("number of member")
plt.plot(df['datetime'], df['num'])
```
