```python
%matplotlib inline
import pandas as pd
```

```python
kernel = kernel[kernel['TotalVotes'] > 0]
```

```python
len(kernel)
```

```python
kernel['Date'] = pd.to_datetime(kernel['MadePublicDate'])
```

```python
kernel2018 =  kernel[kernel['Date'].dt.year == 2018.0].reset_index()
kernel2017 =  kernel[kernel['Date'].dt.year == 2017.0].reset_index()
```

```python
kernel2016 =  kernel[kernel['Date'].dt.year == 2016.0].reset_index()
kernel2015 =  kernel[kernel['Date'].dt.year == 2015.0].reset_index()
```

```python
len(kernel2016)
```

```python
len(kernel2015)
```

```python
kernel2016.head()
```

```python
words2018 = []
words2017 = []
words2016 = []
words2015 = []
```

```python
for _ in (kernel2018['CurrentUrlSlug']):
    words2018 += _.split("-")
```

```python
for _ in (kernel2017['CurrentUrlSlug']):
    words2017 += _.split("-")
```

```python
for _ in (kernel2016['CurrentUrlSlug']):
    try:
        words2016 += _.split("-")
    except:
        pass
```

```python
for _ in (kernel2015['CurrentUrlSlug']):
    words2015 += _.split("-")
```

```python
import collections
c2018 = collections.Counter(words2018)
c2018.most_common()
```

```python
import numpy as np
from sklearn.feature_extraction.text import TfidfVectorizer
from nltk.corpus import stopwords
```

```python
data = [
    ' '.join(words2015),
    ' '.join(words2016),
    ' '.join(words2017),
    ' '.join(words2018),
]
```

```python
stopWords = stopwords.words("english")
```

```python
stopWords.append('data')
stopWords.append('analysis')
stopWords.append('using')
```

```python
vectorizer = TfidfVectorizer(stop_words = stopWords)
X = vectorizer.fit_transform(data).toarray()
```

```python
X.shape
```

```python
df = pd.DataFrame(X.T, index = vectorizer.get_feature_names(), columns = ['words2015', 'words2016', 'words2017', 'words2018'])
```

```python
forplot = df.sort_values('words2018', ascending=False).head(20)
forplot
```

```python
import matplotlib.pyplot as plt
plt.rcParams["font.size"] = 18
forplot.T.plot(figsize=(20, 20))
```
