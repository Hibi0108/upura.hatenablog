```python
import pandas as pd
```

```python
kernel = pd.read_csv('Kernels.csv')
# forum = pd.read_csv('ForumTopics.csv')
```

```python
kernel.head()
```

```python
kernel['Date'] = pd.to_datetime(kernel['MadePublicDate'])
```

```python
kernel2018 =  kernel[kernel['Date'].dt.year == 2018.0].reset_index()
kernel2017 =  kernel[kernel['Date'].dt.year == 2017.0].reset_index()
```

```python
kernel2016 =  kernel[kernel['Date'].dt.year == 2016.0].reset_index()
kernel2015 =  kernel[kernel['Date'].dt.year == 2015.0].reset_index()
```

```python
len(kernel2016)
```

```python
len(kernel2015)
```

```python
kernel2016.head()
```

```python
words2018 = []
words2017 = []
words2016 = []
words2015 = []
```

```python
for _ in (kernel2018['CurrentUrlSlug']):
    words2018 += _.split("-")
```

```python
for _ in (kernel2017['CurrentUrlSlug']):
    words2017 += _.split("-")
```

```python
for _ in (kernel2016['CurrentUrlSlug']):
    try:
        words2016 += _.split("-")
    except:
        pass
```

```python
for _ in (kernel2015['CurrentUrlSlug']):
    words2015 += _.split("-")
```

```python
import collections
c2018 = collections.Counter(words2018)
c2018.most_common()
```

```python
c2017 = collections.Counter(words2017)
c2017.most_common()
```

```python
c2016 = collections.Counter(words2016)
c2016.most_common()
```

```python
c2015 = collections.Counter(words2015)
c2015.most_common()
```

```python
import numpy as np
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
data = [
    ' '.join(words2018),
    ' '.join(words2017),
    ' '.join(words2016),
    ' '.join(words2015),
]
```

```python
vectorizer = TfidfVectorizer()
X = vectorizer.fit_transform(data).toarray()
```

```python
X.shape
```

```python
df = pd.DataFrame(X.T, index = vectorizer.get_feature_names(), columns = ['words2018', 'words2017', 'words2016', 'words2015'])
```
